```python
# Import necessary package
from astropy.io import fits
```

```python
# Make sure the file path is correct and the FITS file is available.
file_dir = 'data/'
img_name = 'pdi_pi_collapsed.fits'
# Load the image data from the FITS file
img = fits.getdata(img_name, ext=1)
```

```python
# Create a figure with a specified size
fig = plt.figure(figsize=(8, 8))

# Display the image using imshow; adjust scaling with vmin and vmax
image = plt.imshow(img, origin='lower', vmin=-0.001, vmax=0.01)

# Add a colorbar to indicate surface brightness values
cbar = fig.colorbar(image, shrink=0.82)
cbar.set_label('Surface Brightness (arb. units)', rotation=90, fontsize=12)
# Set x and y limits to zoom in on the central region of the image
plt.xlim([60, 140])
plt.ylim([60, 140])

# Label the axes using LaTeX formatting for astronomical coordinates
plt.xlabel(r'$\Delta$ RA [px]', fontsize=12)
plt.ylabel(r'$\Delta$ Dec [px]', fontsize=12)

# Add a title to the plot
plt.title('Astronomical FITS Image Visualization')

# Display the figure
plt.show()
# This code loads an astronomical FITS image and visualizes a selected region.
# The colorbar represents surface brightness in arbitrary units.
# Axis labels use common astronomical conventions (Right Ascension and Declination).000001
```

```python
# Import the pandas
import pandas as pd
```

```python
# Read the star catalog (make sure 'closest20kstars.csv' is in the same directory)
stellar = pd.read_csv('closest20kstars.csv')

# Create a matplotlib figure with size 10x10 inches
fig = plt.figure(figsize=[10, 10])

# Plot a scatter plot of Right Ascension (RA) vs Declination (Dec)
plt.scatter(stellar['ra'], stellar['dec'],
            alpha=0.8,       # transparency of points
            s=12,            # marker size
            lw=0.5,          # line width of marker edges
            ec='k',          # edge color (black)
            color='darkorchid')  # marker color

# Set axis limits to match sky coordinates
plt.xlim([0, 362])
plt.ylim([-75, 75])

# Label the axes with units and set the title
plt.xlabel('RA [$\\degree$]', fontsize=14)
plt.ylabel('Dec [$\\degree$]', fontsize=14)
plt.title('Closest 20,000 stars from Gaia Archive (DR2)', fontsize=16)

# Show the figure
plt.show()
# This code reads a CSV catalog of the 20,000 closest stars from the Gaia Archive.
# It creates a scatter plot with Right Ascension (RA) on the x-axis and Declination (Dec) on the y-axis.
# Each point represents a star, colored purple with a black edge.
# The plot visualizes how nearby stars are distributed on the sky.
```

```python
# Assignment3.2: One thing that is important is to know how to save these figures. So why don't you try that?
# Plotting and saving a figure
x = np.linspace(-10, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Example: Sine Wave')
plt.legend()

# Save the figure to a PNG file in your working directory
plt.savefig('sine_wave_example.png')

plt.show()

# The plot above is saved as 'sine_wave_example.png' in the working directory.
```
